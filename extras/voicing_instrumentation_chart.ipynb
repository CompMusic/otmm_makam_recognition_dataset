```python
from fileoperations.fileoperations import get_filenames_in_dir
from collections import Counter
from matplotlib import pyplot
import json
```

```python
data_folder = '../data/'
metadata_files = get_filenames_in_dir(data_folder, keyword='*.json')[0]
```

```python
inst_voice_list = list()
for mf in metadata_files:
    metadata = json.load(open(mf))
    inst_voice_list.append(metadata['instrumentation_voicing'])
```

```python
inst_voice_counter = Counter(inst_voice_list)

pyplot.pie(inst_voice_counter.values(), labels=inst_voice_counter.keys(), autopct='%d', startangle=90)
pyplot.show()
```
